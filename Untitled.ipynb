# Visualising Train Data in different ways

```python
gb_store_sale = TRAIN.groupby(by='store_nbr')['sales'].mean()
plt.figure(figsize=(10,5))
sns.barplot(y=gb_store_sale.values,x=gb_store_sale.index)
plt.xticks(rotation='vertical')
plt.ylabel('sale')
plt.show()
```

```python
gb_family_sale = TRAIN.groupby(by='family')['sales'].mean()
plt.figure(figsize=(10,5))
sns.barplot(y=gb_family_sale.values,x=gb_family_sale.index)
plt.xticks(rotation='vertical')
plt.ylabel('sale')
plt.show()
```
